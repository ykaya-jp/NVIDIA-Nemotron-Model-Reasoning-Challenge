# nvidia-nemotron-model-reasoning-challenge — EDA

1. shape / dtype / missing
2. target distribution
3. train/test distribution diff (kagglib.eda.compare_distributions)
4. correlations


```python
import pandas as pd
from kagglib.data import read_csv_polars
from kagglib.eda import compare_distributions

pd.set_option("display.max_columns", 100)
```

```python
train = read_csv_polars("data/raw/train.csv").to_pandas()
test = read_csv_polars("data/raw/test.csv").to_pandas()
print(train.shape, test.shape)
train.head()
```

```python
drift = compare_distributions(
    train.drop(columns=["target", "id"], errors="ignore"),
    test.drop(columns=["id"], errors="ignore"),
)
drift[drift["drift_flag"]].head(20)
```
